```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid = pd.read_csv('datasets/Provisional_COVID-19_Deaths_by_Sex_and_Age.csv')
covid.head(10)
```

```python
state_region = pd.read_csv('datasets/states.csv')
state_region.head(10)
```

```python
state_population = pd.read_csv('datasets/states_population.csv')
state_population.head(10)
```

```python
temp = pd.merge(covid, state_region, on='State', how='inner')
temp.head(10)
```

```python
df = pd.merge(temp, state_population, on='State', how='inner')
df.head(10)
```

```python
df['State'].unique()
```

```python
pd.value_counts(df['State'].unique()).sum()
```

```python
df.drop(columns=['Footnote', 'Division', 'Data As Of'], inplace=True)
df
```

```python
df = df[df['Year'] != 2023]
df
```

```python
df_allage_allsex = df[(df['Sex'] == 'All Sexes') & (df['Age Group'] == 'All Ages')]
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_allsex, x='State', y='Total Deaths', errorbar=None)
# plt.title(f'Monthly Average {deaths_col} by Year and {sex_col}', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df_allage_eachsex = df[(df['Age Group'] == 'All Ages') & (df['Sex'] != 'All Sexes')]
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_eachsex, x='State', y='Total Deaths', hue='Sex', palette='viridis', errorbar=None)
# plt.title(f'Monthly Average {deaths_col} by Year and {sex_col}', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.legend(title='Sex')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df_eachage_allsex = df[(df['Age Group'] != 'All Ages') & (df['Sex'] == 'All Sexes')]
```

```python
plt.figure(figsize=(30, 6))
sns.barplot(df_eachage_allsex, x='State', y='Total Deaths', hue='Age Group', palette='viridis', errorbar=None)
# plt.title(f'Monthly Average {deaths_col} by Year and {sex_col}', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.legend(title='Sex')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(30, 6))
sns.scatterplot(df_eachage_allsex, x='State', y='Age Group', size='Total Deaths', sizes=(10,100))
# plt.title(f'Monthly Average {deaths_col} by Year and {sex_col}', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.legend(title='Sex')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
