```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid = pd.read_csv('datasets/Provisional_COVID-19_Deaths_by_Sex_and_Age.csv')
covid.head(10)
```

```python
state_region = pd.read_csv('datasets/states.csv')
state_region.head(10)
```

```python
state_population = pd.read_csv('datasets/states_population.csv')
state_population.head(10)
```

```python
temp = pd.merge(covid, state_region, on='State', how='inner')
temp.head(10)
```

```python
df = pd.merge(temp, state_population, on='State', how='inner')
df.head(10)
```

```python
df['State'].unique()
```

```python
pd.value_counts(df['State'].unique()).sum()
```

```python
df.drop(columns=['Footnote', 'Division', 'Data As Of'], inplace=True)
df
```

```python
df = df[df['Year'] != 2023]
df
```

# By Year (2020)

- ## All Sexes and All Ages

```python
df_allage_allsex = df[(df['Sex'] == 'All Sexes') & (df['Age Group'] == 'All Ages') & (df['Group'] == 'By Year') & (df['Year'] == 2020)]
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_allsex, x='State', y='Total Deaths', errorbar=None)
plt.title('Total Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_allsex, x='State', y='COVID-19 Deaths', errorbar=None)
plt.title('COVID-19 Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
covid_percent_df = df_allage_allsex.copy()
covid_percent_df['Percentage'] = (covid_percent_df['COVID-19 Deaths'] / covid_percent_df['Total Deaths'] * 100)

plt.figure(figsize=(20, 6))
sns.barplot(covid_percent_df, x='State', y='Percentage', errorbar=None, color='lightgreen')
plt.title('Percentage of COVID-19 Deaths Relative to Total Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Percentage %', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_allsex, x='State', y='Influenza Deaths', errorbar=None)
plt.title('Influenza Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Influenza Deaths', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
influenza_percent = df_allage_allsex.copy()
influenza_percent['Percentage'] = (influenza_percent['Influenza Deaths'] / influenza_percent['Total Deaths'] * 100)

plt.figure(figsize=(20, 6))
sns.barplot(influenza_percent, x='State', y='Percentage', errorbar=None, color='lightgreen')
plt.title('Percentage of Influenza Deaths Relative to Total Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Percentage %', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(df_allage_allsex, x='State', y='Pneumonia Deaths', errorbar=None)
plt.title('Pneumonia Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Pneumonia Deaths', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
pneumonia_percent = df_allage_allsex.copy()
pneumonia_percent['Percentage'] = (pneumonia_percent['Pneumonia Deaths'] / pneumonia_percent['Total Deaths'] * 100)

plt.figure(figsize=(20, 6))
sns.barplot(pneumonia_percent, x='State', y='Percentage', errorbar=None, color='lightgreen')
plt.title('Percentage of Pneumonia Deaths Relative to Total Deaths each states (2020)', fontsize=20)
plt.xlabel('State', fontsize=12)
plt.ylabel('Percentage %', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
